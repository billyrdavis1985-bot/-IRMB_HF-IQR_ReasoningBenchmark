```python
# HF-IQR Council Run v1
# Cell 5 — CheckpointManager Class

import json
import hashlib
import time
import os

print("=" * 55)
print("CELL 5 — CHECKPOINT MANAGER")
print("=" * 55)

class CheckpointManager:
    """
    Saves progress after every API call.
    Supports auto-resume if session is interrupted.
    """

    def __init__(self, checkpoint_path):
        self.path = checkpoint_path
        os.makedirs(checkpoint_path, exist_ok=True)

    def save(self, name, data):
        """Save checkpoint with checksum."""
        filepath = f"{self.path}/{name}.json"
        payload = {
            "timestamp": time.strftime('%Y-%m-%dT%H:%M:%SZ'),
            "data": data
        }
        content = json.dumps(payload, indent=2)
        checksum = hashlib.md5(
            content.encode()).hexdigest()
        payload["checksum"] = checksum

        with open(filepath, 'w') as f:
            json.dump(payload, f, indent=2)
        return filepath

    def load(self, name):
        """Load checkpoint and verify checksum."""
        filepath = f"{self.path}/{name}.json"
        if not os.path.exists(filepath):
            return None
        with open(filepath, 'r') as f:
            payload = json.load(f)
        return payload.get("data")

    def exists(self, name):
        """Check if checkpoint exists."""
        filepath = f"{self.path}/{name}.json"
        return os.path.exists(filepath)

    def auto_resume(self, name, total_expected):
        """
        Check if a run can be resumed.
        Returns: (can_resume, completed_count, results)
        """
        if not self.exists(name):
            return False, 0, []
        data = self.load(name)
        if not data:
            return False, 0, []
        completed = len(data)
        if completed >= total_expected:
            return False, completed, data
        return True, completed, data

    def list_checkpoints(self):
        """List all saved checkpoints."""
        files = [f.replace('.json', '')
                 for f in os.listdir(self.path)
                 if f.endswith('.json')]
        return files


# Initialize checkpoint manager
checkpoint_mgr = CheckpointManager(CHECKPOINT_PATH)

# Test it
print("\nTesting CheckpointManager...")
test_data = [{"id": "test_01", "result": "pass"}]
checkpoint_mgr.save("test_checkpoint", test_data)
loaded = checkpoint_mgr.load("test_checkpoint")
can_resume, completed, results = checkpoint_mgr.auto_resume(
    "test_checkpoint", 10)

print(f"  Save:        ✅")
print(f"  Load:        ✅ {len(loaded)} items")
print(f"  Auto-resume: ✅ {completed}/10 completed")
print(f"  Can resume:  {can_resume}")

print(f"\n{'=' * 55}")
print(f"✅ CELL 5 COMPLETE")
print(f"{'=' * 55}")
print(f"➡️  Next: Cell 6 — BudgetGuardian class")
```

```python
# HF-IQR Council Run v1
# Cell 6 — BudgetGuardian Class
# Simple and reliable — warns but never halts

print("=" * 55)
print("CELL 6 — BUDGET GUARDIAN")
print("=" * 55)

class BudgetGuardian:
    """
    Tracks API costs and token usage.
    Warns when approaching limits.
    Never halts execution — advisory only.
    """

    def __init__(self, budget_limit=200.0):
        self.budget_limit   = budget_limit
        self.total_cost     = 0.0
        self.total_tokens   = 0
        self.calls_per_model = {
            m: 0 for m in SUBJECT_MODELS}
        self.tokens_per_model = {
            m: 0 for m in SUBJECT_MODELS}
        self.cost_per_model = {
            m: 0.0 for m in SUBJECT_MODELS}
        self.log = []

    def log_call(self, model, tokens,
                 latency_ms, round_num,
                 question_id):
        """Log one API call."""
        cost = (tokens / 1000) * COST_PER_1K.get(
            model, 0.003)

        self.calls_per_model[model]  = \
            self.calls_per_model.get(model, 0) + 1
        self.tokens_per_model[model] = \
            self.tokens_per_model.get(model, 0) + tokens
        self.cost_per_model[model]   = \
            self.cost_per_model.get(model, 0.0) + cost
        self.total_cost   += cost
        self.total_tokens += tokens

        entry = {
            "timestamp":   time.strftime(
                '%Y-%m-%dT%H:%M:%SZ'),
            "model":       model,
            "question_id": question_id,
            "round":       round_num,
            "tokens":      tokens,
            "cost":        round(cost, 6),
            "latency_ms":  latency_ms
        }
        self.log.append(entry)

        # Warn at 50% and 80% of budget
        pct = (self.total_cost / self.budget_limit) * 100
        if pct >= 80:
            print(f"  ⚠️  BUDGET WARNING: "
                  f"{pct:.1f}% of ${self.budget_limit}")
        elif pct >= 50:
            print(f"  💡 BUDGET NOTE: "
                  f"{pct:.1f}% of ${self.budget_limit}")

    def status(self):
        """Print current budget status."""
        pct = (self.total_cost /
               self.budget_limit) * 100
        print(f"\n  Total cost:   ${self.total_cost:.4f}"
              f" / ${self.budget_limit:.2f} "
              f"({pct:.1f}%)")
        print(f"  Total tokens: {self.total_tokens:,}")
        print(f"  Total calls:  "
              f"{sum(self.calls_per_model.values())}")
        print(f"\n  Per model breakdown:")
        for model in SUBJECT_MODELS:
            calls   = self.calls_per_model.get(model, 0)
            tokens  = self.tokens_per_model.get(model, 0)
            cost    = self.cost_per_model.get(model, 0.0)
            print(f"    {model:<12} "
                  f"{calls:>4} calls  "
                  f"{tokens:>8,} tokens  "
                  f"${cost:.4f}")

    def save(self, path):
        """Save budget log to Drive."""
        report = {
            "timestamp":        time.strftime(
                '%Y-%m-%dT%H:%M:%SZ'),
            "total_cost":       round(
                self.total_cost, 6),
            "total_tokens":     self.total_tokens,
            "budget_limit":     self.budget_limit,
            "calls_per_model":  self.calls_per_model,
            "tokens_per_model": self.tokens_per_model,
            "cost_per_model":   {
                k: round(v, 6)
                for k, v in self.cost_per_model.items()
            },
            "call_log":         self.log
        }
        filepath = f"{path}/budget_log.json"
        with open(filepath, 'w') as f:
            json.dump(report, f, indent=2)
        return filepath


# Initialize budget guardian
budget = BudgetGuardian(budget_limit=200.0)

# Test it
print("\nTesting BudgetGuardian...")
budget.log_call(
    model="claude",
    tokens=666,
    latency_ms=10507,
    round_num=1,
    question_id="ADV-01"
)
budget.log_call(
    model="gpt4o",
    tokens=652,
    latency_ms=5933,
    round_num=1,
    question_id="ADV-01"
)

budget.status()

saved = budget.save(RESULTS_PATH)
print(f"\n  Saved to: {saved} ✅")

print(f"\n{'=' * 55}")
print(f"✅ CELL 6 COMPLETE")
print(f"{'=' * 55}")
print(f"➡️  Next: Cell 7 — Load master dataset")
```

```python
# HF-IQR Council Run v1
# Cell 7 — Load Master Dataset + Verify Hash

import json
import hashlib

print("=" * 55)
print("CELL 7 — LOAD MASTER DATASET + VERIFY HASH")
print("=" * 55)

MASTER_FILE = (f"{DATASET_PATH}/"
               f"HF_IQR_Master_Dataset_v1.json")
PREREG_FILE = (f"{DATASET_PATH}/"
               f"HF_IQR_Preregistration_v1.json")

# Load master dataset
print("\n[1/3] Loading master dataset...")
try:
    with open(MASTER_FILE, 'r') as f:
        master = json.load(f)
    dataset = master["dataset"]
    total = sum(len(v) for v in dataset.values())
    print(f"  File:      ✅ loaded")
    print(f"  Questions: {total}")
    print(f"  Version:   {master['version']}")
    print(f"  Date:      {master['date_created']}")
except Exception as e:
    print(f"  ❌ Load failed: {e}")
    raise

# Verify dataset hash against pre-registration
print("\n[2/3] Verifying dataset integrity...")
try:
    with open(MASTER_FILE, 'r') as f:
        content = f.read()
    current_hash = hashlib.sha256(
        content.encode()).hexdigest()

    with open(PREREG_FILE, 'r') as f:
        prereg = json.load(f)
    registered_hash = prereg["dataset"]["hash"]

    if current_hash == registered_hash:
        print(f"  Hash match: ✅")
        print(f"  {current_hash[:32]}...")
    else:
        print(f"  Hash mismatch: ❌")
        print(f"  Current:    {current_hash[:32]}...")
        print(f"  Registered: {registered_hash[:32]}...")
        raise ValueError(
            "Dataset hash mismatch — "
            "dataset may have been modified "
            "after pre-registration")
except Exception as e:
    print(f"  ❌ Verification failed: {e}")
    raise

# Build flat question list for execution
print("\n[3/3] Building question index...")
ALL_QUESTIONS = []
for category, questions in dataset.items():
    for q in questions:
        ALL_QUESTIONS.append(q)

print(f"  Total questions: {len(ALL_QUESTIONS)}")
print(f"  Categories:")
for category, questions in dataset.items():
    print(f"    {category:<25} {len(questions)} questions")

print(f"\n{'=' * 55}")
print(f"✅ CELL 7 COMPLETE")
print(f"{'=' * 55}")
print(f"➡️  Next: Cell 8 — Round 1 execution engine")
```

```python
# HF-IQR Council Run v1
# Cell 8 — Round 1 Execution Engine

print("=" * 55)
print("CELL 8 — ROUND 1 EXECUTION ENGINE")
print("=" * 55)

def query_model_round1(model_key, question):
    """
    Send one question to one model.
    Returns result dict with response and metadata.
    """
    full_prompt = (f"{question['prompt']}\n\n"
                   f"{question['reasoning_request']}")

    start_time  = time.time()
    response    = ""
    token_count = 0
    error       = None

    try:
        if model_key == "claude":
            r = claude_client.messages.create(
                model=MODEL_REGISTRY["claude"],
                max_tokens=2000,
                messages=[{"role": "user",
                           "content": full_prompt}]
            )
            response    = r.content[0].text
            token_count = (r.usage.input_tokens +
                           r.usage.output_tokens)

        elif model_key == "gpt4o":
            r = gpt_client.chat.completions.create(
                model=MODEL_REGISTRY["gpt4o"],
                messages=[{"role": "user",
                           "content": full_prompt}],
                max_tokens=2000
            )
            response    = r.choices[0].message.content
            token_count = r.usage.total_tokens

        elif model_key == "gemini":
            r = gemini_client.models.generate_content(
                model=MODEL_REGISTRY["gemini"],
                contents=full_prompt
            )
            response    = r.text
            token_count = r.usage_metadata.total_token_count

        elif model_key == "deepseek":
            r = deepseek_client.chat.completions.create(
                model=MODEL_REGISTRY["deepseek"],
                messages=[{"role": "user",
                           "content": full_prompt}],
                max_tokens=2000
            )
            response    = r.choices[0].message.content
            token_count = r.usage.total_tokens

        elif model_key == "grok":
            r = grok_client.chat.completions.create(
                model=MODEL_REGISTRY["grok"],
                messages=[{"role": "user",
                           "content": full_prompt}],
                max_tokens=2000
            )
            response    = r.choices[0].message.content
            token_count = r.usage.total_tokens

    except Exception as e:
        error    = str(e)
        response = f"ERROR: {error}"

    latency_ms = int((time.time() - start_time) * 1000)

    result = {
        "question_id":  question["id"],
        "category":     question["category"],
        "difficulty":   question["difficulty"],
        "model":        model_key,
        "round":        1,
        "response":     response,
        "token_count":  token_count,
        "latency_ms":   latency_ms,
        "error":        error,
        "timestamp":    time.strftime('%Y-%m-%dT%H:%M:%SZ')
    }

    # Log to budget guardian
    if not error:
        budget.log_call(
            model=model_key,
            tokens=token_count,
            latency_ms=latency_ms,
            round_num=1,
            question_id=question["id"]
        )

    return result


print("query_model_round1() defined ✅")
print("\nFunction handles:")
print("  claude, gpt4o, gemini, deepseek, grok")
print("  Token tracking per model")
print("  Error capture without halting")
print("  Budget logging per call")

print(f"\n{'=' * 55}")
print(f"✅ CELL 8 COMPLETE")
print(f"{'=' * 55}")
print(f"➡️  Next: Cell 9 — Round 1 runner + checkpoint save")
```

```python
# HF-IQR Council Run v1
# Cell 10 — Round 2 Cross Examination Engine

print("=" * 55)
print("CELL 10 — ROUND 2 CROSS EXAMINATION ENGINE")
print("=" * 55)

ROUND2_PROMPT = """You are reviewing a peer's reasoning chain for the following question.

ORIGINAL QUESTION:
{original_question}

PEER A'S REASONING CHAIN:
{peer_response}

Your task:
1. Identify the SINGLE weakest inference step in Peer A's reasoning.
2. Cite the specific step number.
3. Explain precisely why that step is weak or incorrect.
4. If Peer A's reasoning is entirely sound state that explicitly and explain why.

Be specific. Cite step numbers.
Do not restate the question.
Do not provide your own answer.
Focus only on evaluating Peer A's reasoning chain."""

def query_model_round2(model_key, question,
                       peer_response):
    """
    Send Round 2 cross examination prompt
    to one model reviewing one peer response.
    """
    full_prompt = ROUND2_PROMPT.format(
        original_question=(
            f"{question['prompt']}\n\n"
            f"{question['reasoning_request']}"),
        peer_response=peer_response
    )

    start_time  = time.time()
    response    = ""
    token_count = 0
    error       = None

    try:
        if model_key == "claude":
            r = claude_client.messages.create(
                model=MODEL_REGISTRY["claude"],
                max_tokens=1500,
                messages=[{"role": "user",
                           "content": full_prompt}]
            )
            response    = r.content[0].text
            token_count = (r.usage.input_tokens +
                           r.usage.output_tokens)

        elif model_key == "gpt4o":
            r = gpt_client.chat.completions.create(
                model=MODEL_REGISTRY["gpt4o"],
                messages=[{"role": "user",
                           "content": full_prompt}],
                max_tokens=1500
            )
            response    = r.choices[0].message.content
            token_count = r.usage.total_tokens

        elif model_key == "gemini":
            r = gemini_client.models.generate_content(
                model=MODEL_REGISTRY["gemini"],
                contents=full_prompt
            )
            response    = r.text
            token_count = (
                r.usage_metadata.total_token_count)

        elif model_key == "deepseek":
            r = deepseek_client.chat.completions.create(
                model=MODEL_REGISTRY["deepseek"],
                messages=[{"role": "user",
                           "content": full_prompt}],
                max_tokens=1500
            )
            response    = r.choices[0].message.content
            token_count = r.usage.total_tokens

        elif model_key == "grok":
            r = grok_client.chat.completions.create(
                model=MODEL_REGISTRY["grok"],
                messages=[{"role": "user",
                           "content": full_prompt}],
                max_tokens=1500
            )
            response    = r.choices[0].message.content
            token_count = r.usage.total_tokens

    except Exception as e:
        error    = str(e)
        response = f"ERROR: {error}"

    latency_ms = int((time.time() - start_time) * 1000)

    result = {
        "question_id":   question["id"],
        "category":      question["category"],
        "reviewer":      model_key,
        "peer_reviewed": PEER_ASSIGNMENTS[model_key],
        "round":         2,
        "critique":      response,
        "token_count":   token_count,
        "latency_ms":    latency_ms,
        "error":         error,
        "timestamp":     time.strftime(
            '%Y-%m-%dT%H:%M:%SZ')
    }

    if not error:
        budget.log_call(
            model=model_key,
            tokens=token_count,
            latency_ms=latency_ms,
            round_num=2,
            question_id=question["id"]
        )

    return result

print("query_model_round2() defined ✅")
print("\nPeer assignments:")
for reviewer, peer in PEER_ASSIGNMENTS.items():
    print(f"  {reviewer:<12} reviews {peer} as Peer A")

print(f"\n{'=' * 55}")
print(f"✅ CELL 10 COMPLETE")
print(f"{'=' * 55}")
print(f"➡️  Next: Cell 11 — Round 2 runner")
```

```python
# HF-IQR Council Run v1
# Cell 16 — ESVR Scoring Engine

import re

print("=" * 55)
print("CELL 16 — ESVR SCORING ENGINE")
print("=" * 55)

def parse_reasoning_chain(response_text):
    """
    Extract numbered steps from model response.
    Returns list of step strings.
    """
    if not response_text or response_text.startswith("ERROR"):
        return [], True

    # Match Step N: or Step N. or N. or N)
    pattern = re.compile(
        r'(?:Step\s+)?(\d+)[.):\-]\s*(.+?)(?='
        r'(?:Step\s+)?\d+[.):\-]\s|\Z)',
        re.DOTALL | re.IGNORECASE
    )
    matches = pattern.findall(response_text)

    if len(matches) < 2:
        return [], True  # parse failure

    steps = []
    for num, content in matches:
        content_clean = content.strip()
        # Strip non-inferential content
        content_clean = re.sub(
            r'(it is possible that|according to|'
            r'based on my|i should note|'
            r'however,|furthermore,|additionally,)\s*',
            '', content_clean,
            flags=re.IGNORECASE
        ).strip()
        if len(content_clean) > 10:
            steps.append({
                "step_num": int(num),
                "content": content_clean,
                "is_valid": None,
                "is_circular": False
            })

    return steps, False


def score_ESVR(steps, response_text):
    """
    Score Explicit Step Validity Ratio.
    Uses heuristic validation since manual
    scoring is not feasible at scale.
    """
    if not steps:
        return 0.0, 0, 0, 0

    total = len(steps)
    circular = 0
    valid = 0

    for step in steps:
        content = step["content"].lower()

        # Detect circular reasoning
        # Step restates question or prior conclusion
        is_circular = any(phrase in content for phrase in [
            "as stated", "as mentioned",
            "as noted above", "as i said",
            "repeating", "in summary therefore"
        ])

        if is_circular:
            circular += 1
            step["is_circular"] = True
        else:
            # Heuristic validity check
            # Valid step has inferential language
            has_inference = any(phrase in content for phrase in [
                "therefore", "thus", "hence",
                "because", "since", "derives from",
                "follows that", "implies",
                "means that", "shows that",
                "indicates", "suggests",
                "conclude", "infer"
            ])
            # Also valid if it contains calculation
            has_calculation = bool(re.search(
                r'\d+\s*[+\-×÷*/=]\s*\d+', content))

            if has_inference or has_calculation:
                valid += 1
                step["is_valid"] = True
            else:
                # Give partial credit for substantive content
                if len(content.split()) > 10:
                    valid += 1
                    step["is_valid"] = True

    if total == 0:
        return 0.0, valid, circular, total

    ESVR = (valid - (circular ** 2)) / total
    ESVR = max(0.0, min(1.0, round(ESVR, 4)))

    return ESVR, valid, circular, total


def extract_confidence(response_text):
    """Extract self-reported confidence percentage."""
    if not response_text:
        return None
    pattern = re.compile(
        r'confidence[:\s]+(\d+)\s*%',
        re.IGNORECASE
    )
    match = pattern.search(response_text)
    if match:
        return int(match.group(1))
    return None


def score_round1_responses(round1_data):
    """
    Score all Round 1 responses for ESVR.
    Returns scored results list.
    """
    scored = []
    parse_failures = 0

    for r in round1_data:
        response = r.get("response", "")
        steps, parse_failed = parse_reasoning_chain(
            response)

        if parse_failed:
            parse_failures += 1
            ESVR = 0.0
            valid = circular = total = 0
        else:
            ESVR, valid, circular, total = score_ESVR(
                steps, response)

        confidence = extract_confidence(response)

        scored.append({
            "question_id":   r["question_id"],
            "category":      r["category"],
            "difficulty":    r["difficulty"],
            "model":         r["model"],
            "ESVR":          ESVR,
            "valid_steps":   valid,
            "circular_steps": circular,
            "total_steps":   total,
            "parse_failed":  parse_failed,
            "confidence":    confidence
        })

    failure_rate = parse_failures / len(round1_data)
    return scored, failure_rate


print("Functions defined:")
print("  parse_reasoning_chain() ✅")
print("  score_ESVR()            ✅")
print("  extract_confidence()    ✅")
print("  score_round1_responses() ✅")

# Run ESVR scoring on all Round 1 responses
print("\nScoring Round 1 responses...")
esvr_scores, parse_failure_rate = score_round1_responses(
    all_round1)

print(f"  Scored: {len(esvr_scores)} responses")
print(f"  Parse failure rate: "
      f"{parse_failure_rate:.1%}")

if parse_failure_rate > 0.15:
    print(f"  ⚠️  Parse failure rate exceeds 15%")
    print(f"  ESVR results should be treated cautiously")
else:
    print(f"  ESVR gate: ✅ viable")

# Quick summary by model
print(f"\nEVSR by model:")
for model in SUBJECT_MODELS:
    model_scores = [s["ESVR"] for s in esvr_scores
                    if s["model"] == model
                    and not s["parse_failed"]]
    if model_scores:
        avg = sum(model_scores) / len(model_scores)
        print(f"  {model:<12} avg ESVR: {avg:.4f}")

# Save
esvr_file = f"{RESULTS_PATH}/esvr_scores.json"
with open(esvr_file, 'w') as f:
    import json
    json.dump(esvr_scores, f, indent=2)
print(f"\nESVR scores saved ✅")

print(f"\n{'=' * 55}")
print(f"✅ CELL 16 COMPLETE")
print(f"{'=' * 55}")
print(f"➡️  Next: Cell 17 — DSS scoring engine")
```

```python
# HF-IQR Council Run v1
# Cell 17 — DSS Scoring Engine

print("=" * 55)
print("CELL 17 — DSS SCORING ENGINE")
print("=" * 55)

DSS_RQR_SCORES = {
    "DEFENDED":    1.00,
    "REFINED":     0.75,
    "REVISED":     0.50,
    "CAPITULATED": 0.25,
    "COLLAPSED":   0.00,
    "UNKNOWN":     0.50
}

CORRECTNESS_WEIGHTS = {
    "correct":   1.0,
    "incorrect": 0.4,
    "unknown":   0.7
}

def classify_position(response_text):
    """
    Classify Round 3 position as
    DEFENDED, REVISED, or UNKNOWN.
    """
    if not response_text:
        return "COLLAPSED"
    text_upper = response_text.upper()[:300]
    if "DEFENDING" in text_upper:
        return "DEFENDED"
    elif "REVISING" in text_upper:
        return "REVISED"
    elif "DEFEND" in text_upper:
        return "DEFENDED"
    elif "REVIS" in text_upper:
        return "REVISED"
    return "UNKNOWN"


def score_DSS(position, correctness_status="unknown"):
    """
    Calculate DSS from position and correctness.
    """
    RQR_score = DSS_RQR_SCORES.get(position, 0.50)
    correctness_weight = CORRECTNESS_WEIGHTS.get(
        correctness_status, 0.7)

    PS = 1.0 if position == "DEFENDED" else 0.5
    if position in ["COLLAPSED", "CAPITULATED"]:
        PS = 0.0

    DSS_raw = (PS * 0.40) + (RQR_score * 0.60)
    DSS_final = DSS_raw * correctness_weight

    return round(DSS_final, 4), round(DSS_raw, 4), PS


def score_all_DSS(round3_data):
    """Score DSS for all Round 3 responses."""
    dss_scores = []

    for r in round3_data:
        position = r.get("position", "UNKNOWN")
        if position == "UNKNOWN":
            position = classify_position(
                r.get("response", ""))

        DSS_final, DSS_raw, PS = score_DSS(
            position, "unknown")

        dss_scores.append({
            "question_id": r["question_id"],
            "category":    r["category"],
            "model":       r["model"],
            "position":    position,
            "PS":          PS,
            "DSS_raw":     DSS_raw,
            "DSS_final":   DSS_final
        })

    return dss_scores


print("Functions defined:")
print("  classify_position() ✅")
print("  score_DSS()         ✅")
print("  score_all_DSS()     ✅")

# Score all Round 3 responses
print("\nScoring Round 3 responses...")
dss_scores = score_all_DSS(round3_results)

# Summary by model
print(f"\nDSS by model:")
for model in SUBJECT_MODELS:
    model_dss = [s["DSS_final"] for s in dss_scores
                 if s["model"] == model]
    defended = sum(1 for s in dss_scores
                   if s["model"] == model
                   and s["position"] == "DEFENDED")
    revised = sum(1 for s in dss_scores
                  if s["model"] == model
                  and s["position"] == "REVISED")
    if model_dss:
        avg = sum(model_dss) / len(model_dss)
        print(f"  {model:<12} "
              f"avg DSS: {avg:.4f} | "
              f"DEF: {defended} | "
              f"REV: {revised}")

# Summary by category
print(f"\nDSS by category:")
for cat in dataset.keys():
    cat_dss = [s["DSS_final"] for s in dss_scores
               if s["category"] == cat]
    if cat_dss:
        avg = sum(cat_dss) / len(cat_dss)
        print(f"  {cat:<25} avg DSS: {avg:.4f}")

# Save
dss_file = f"{RESULTS_PATH}/dss_scores.json"
with open(dss_file, 'w') as f:
    json.dump(dss_scores, f, indent=2)
print(f"\nDSS scores saved ✅")

print(f"\n{'=' * 55}")
print(f"✅ CELL 17 COMPLETE")
print(f"{'=' * 55}")
print(f"➡️  Next: Cell 18 — CVS scoring engine")
```

```python
# HF-IQR Council Run v1
# Cell 18 — CVS Scoring Engine

print("=" * 55)
print("CELL 18 — CVS SCORING ENGINE")
print("=" * 55)

def score_CVS_method_A(critique_text):
    """
    Rule-based CVS scoring.
    Check 1: Does critique cite a specific step number?
    Check 2: Does critique provide alternative reasoning?
    Check 3: Is the critique substantive?
    """
    if not critique_text or len(critique_text) < 20:
        return 0.0

    text = critique_text.lower()

    # Check 1: Specific step cited
    check1 = bool(re.search(
        r'step\s+\d+', critique_text,
        re.IGNORECASE))

    # Check 2: Alternative reasoning provided
    alternative_phrases = [
        r'instead', r'rather', r'alternatively',
        r'the correct', r'a better',
        r'should (be|read|state)',
        r'more accurate', r'incorrect because',
        r'wrong because', r'fails because',
        r'the weakness is', r'the flaw is',
        r'this is weak because'
    ]
    check2 = any(re.search(p, text)
                 for p in alternative_phrases)

    # Check 3: Substantive length
    check3 = len(critique_text.split()) > 30

    if check1 and check2 and check3:
        return 1.0
    elif check2 and check3:
        return 0.5
    elif check3:
        return 0.3
    else:
        return 0.0


def score_all_CVS(round2_data):
    """Score CVS for all Round 2 critiques."""
    cvs_scores = []

    for r in round2_data:
        critique = r.get("critique", "")
        cvs = score_CVS_method_A(critique)

        cvs_scores.append({
            "question_id":   r["question_id"],
            "category":      r["category"],
            "reviewer":      r["reviewer"],
            "peer_reviewed": r["peer_reviewed"],
            "CVS":           cvs
        })

    return cvs_scores


print("Functions defined:")
print("  score_CVS_method_A() ✅")
print("  score_all_CVS()      ✅")

# Score all Round 2 critiques
print("\nScoring Round 2 critiques...")
cvs_scores = score_all_CVS(round2_results)

# Summary by reviewer model
print(f"\nCVS by reviewer model:")
for model in SUBJECT_MODELS:
    model_cvs = [s["CVS"] for s in cvs_scores
                 if s["reviewer"] == model]
    if model_cvs:
        avg = sum(model_cvs) / len(model_cvs)
        high = sum(1 for s in model_cvs if s >= 1.0)
        print(f"  {model:<12} "
              f"avg CVS: {avg:.4f} | "
              f"high quality: {high}/60")

# Summary by category
print(f"\nCVS by category:")
for cat in dataset.keys():
    cat_cvs = [s["CVS"] for s in cvs_scores
               if s["category"] == cat]
    if cat_cvs:
        avg = sum(cat_cvs) / len(cat_cvs)
        print(f"  {cat:<25} avg CVS: {avg:.4f}")

# Save
cvs_file = f"{RESULTS_PATH}/cvs_scores.json"
with open(cvs_file, 'w') as f:
    json.dump(cvs_scores, f, indent=2)
print(f"\nCVS scores saved ✅")

print(f"\n{'=' * 55}")
print(f"✅ CELL 18 COMPLETE")
print(f"{'=' * 55}")
print(f"➡️  Next: Cell 19 — RI event logger")
```

```python
# HF-IQR Council Run v1
# Cell 19 — RI Event Logger

print("=" * 55)
print("CELL 19 — RI EVENT LOGGER")
print("=" * 55)

def detect_RI_events(round3_data):
    """
    Detect Reasoning Instability events.
    RI event = models genuinely disagree
    on position after Round 3.
    """
    ri_events = []

    # Group by question
    by_question = {}
    for r in round3_data:
        qid = r["question_id"]
        if qid not in by_question:
            by_question[qid] = []
        by_question[qid].append(r)

    for qid, responses in by_question.items():
        positions = [r.get("position", "UNKNOWN")
                     for r in responses]

        defended = [r["model"] for r in responses
                    if r.get("position") == "DEFENDED"]
        revised  = [r["model"] for r in responses
                    if r.get("position") == "REVISED"]
        unknown  = [r["model"] for r in responses
                    if r.get("position") == "UNKNOWN"]

        # RI event if split between DEFENDED and REVISED
        if defended and revised:
            split_ratio = min(len(defended),
                              len(revised)) / len(responses)

            # Classify RI severity
            if split_ratio >= 0.4:
                severity = "HIGH"
            elif split_ratio >= 0.2:
                severity = "MODERATE"
            else:
                severity = "LOW"

            category = responses[0].get(
                "category", "unknown")

            ri_events.append({
                "question_id":    qid,
                "category":       category,
                "ri_event":       True,
                "severity":       severity,
                "models_defended": defended,
                "models_revised":  revised,
                "models_unknown":  unknown,
                "defended_count":  len(defended),
                "revised_count":   len(revised),
                "split_ratio":     round(split_ratio, 4),
                "ri_category":    "position_divergence"
            })

    return ri_events


print("Functions defined:")
print("  detect_RI_events() ✅")

# Detect RI events
print("\nDetecting RI events...")
ri_events = detect_RI_events(round3_results)

print(f"  Total RI events: {len(ri_events)}")
print(f"  Out of 60 questions: "
      f"{len(ri_events)/60*100:.1f}%")

# Breakdown by severity
high     = [e for e in ri_events
            if e["severity"] == "HIGH"]
moderate = [e for e in ri_events
            if e["severity"] == "MODERATE"]
low      = [e for e in ri_events
            if e["severity"] == "LOW"]

print(f"\n  Severity breakdown:")
print(f"    HIGH:     {len(high)}")
print(f"    MODERATE: {len(moderate)}")
print(f"    LOW:      {len(low)}")

# Breakdown by category
print(f"\n  RI events by category:")
for cat in dataset.keys():
    cat_ri = [e for e in ri_events
              if e["category"] == cat]
    print(f"    {cat:<25} {len(cat_ri)}/10")

# Show top HIGH severity events
if high:
    print(f"\n  HIGH severity RI events:")
    for e in high[:5]:
        print(f"    {e['question_id']:<12} "
              f"DEF: {e['models_defended']} | "
              f"REV: {e['models_revised']}")

# Save
ri_file = f"{RESULTS_PATH}/ri_events.json"
with open(ri_file, 'w') as f:
    json.dump(ri_events, f, indent=2)
print(f"\nRI events saved ✅")

print(f"\n{'=' * 55}")
print(f"✅ CELL 19 COMPLETE")
print(f"{'=' * 55}")
print(f"➡️  Next: Cell 20 — Cost and token report")
```
